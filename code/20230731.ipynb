# 경사하강법

```python
# 지도학습: 문제집, 정답지

import numpy as np #병렬연산
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8]) #공부시간
x2 = np.array([1, 2, 1, 3]) #문제집 푼 개수
y = np.array([20,55,45,80])

a1 = 0
a2 = 0
b = 0
lr = 0.005

for n in range(3000):
    y_hat = a1 * x1  + a2 * x2 + b
    error = y_hat - y

    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)

    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr

    print(f"{n+1}회 학습 // a1 : {a1_diff} a2 : {a2_diff} // b : {b_diff}")
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1 *  x1 + a2 * x2 + b, 'red')
```

```python
a1 * 5 + a2 * 2 + b
```

```python
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

a = 0
b = 0
c = 0
d = 0
lr = 0.0000025

for n in range(5000):
    y_hat = a * x ** 3 + b * x ** 2 + c * x + d 
    error = y_hat - y

    a_diff = sum(2 * error * x ** 3)
    b_diff = sum(2 * error * x ** 2)
    c_diff = sum(2 * error * x)
    d_diff = sum(2 * error * 1)

    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr

    print(f"{n+1}회 학습 // a : {a_diff: .1f}  // b : {b_diff: .1f} // c : {c_diff: .1f} // d : {d_diff: .1f}")
```

```python
plt.plot(x,y)
plt.plot(x, a * x ** 3 + b * x ** 2 + c * x + d, 'red')
```

# 지도학습(분류) : 범주형 변수를 예측

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fish = pd.read_csv("data/fish.csv") #길이, 무게 -> 도미, 빙어
fish
```

```python
#빙어 <= 15cm, < 20g
#도미 > 25cm, 240g >

sns.lmplot(data = fish, x = 'length', y = 'weight', hue='target',
           fit_reg=False)
```

```python
# 최근접 이웃 /  유사한 특성을 가진 데이터는 유사한 범주에 속하는 경향이 있다는 가정하에 사용
from sklearn.neighbors import KNeighborsClassifier

data = fish[['length', 'weight']]
target = fish['target']

data = np.array(data)
target = np.array(target)

model = KNeighborsClassifier() #모델생성
model.fit(data,target) #학습
```

```python
box = [40, 400]
model.predict([box]) #유클리디언 거리 계산 방식 / 예측
```

```python
# 훈련데이터 / 시험데이터 분할
# sklearn : 75 : 25(권장)
# google : 8 : 2 (권장) //최대한 많이 훈련데이터에 쏟아야함

# 잘 섞어서 비율까지 맞춰서 분할
#데이터 분할을 위한 함수
from sklearn.model_selection import train_test_split

# 데이터를 모두 집어넣어 분석해 모델을 만들 수 있겠지만, 새로운 데이터가 나타나야 그 모델을 테스트 할 수 있다.
# 그러므로 모델을 만들때 일부를 남겨두고 모델을 만든다음, 그 남겨진 데이터를 테스트용 데이터로서 사용한다.
# train은 훈련용, test 테스트용 75%, 25% 섞어서 랜덤하게(뽑는 방법을 설정할수도 있다.)
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2) # 시험 데이터에 20%를 할당한다
```

```python
plt.scatter(train_input[:,0], train_input[:, 1])
plt.scatter(test_input[:,0], test_input[:, 1])
plt.legend(['train', 'test'])
```

```python
model = KNeighborsClassifier()
model.fit(train_input, train_target)
```

```python
from sklearn.neighbors import KNeighborsRegressor
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

perch = pd.read_csv('data/perch.csv') #무게를 예측

data = perch.iloc[:, :-1].to_numpy()
target = perch['weight'].to_numpy()

#75:25
train_input, test_input, train_target, test_target = train_test_split(data, target)

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)
#mean = np.mean(train_input, axis=0)
#std = np.std(train_input, axis=0)
#train_scaled = (train_input - mean) / std
#test_scaled = (test_input - mean) / std

model = KNeighborsRegressor()
model.fit(train_scaled, train_target)
model.score(test_scaled, test_target) #R^2 : 예측값이 실제값과 얼마나 비슷한가?



```
